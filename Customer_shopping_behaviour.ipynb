```python
#create a columsn of age group
def category(age):
  if age<=25:
    return "Young Adult"
  elif age<=35:
    return "Adult"
  elif age<=50:
    return "Middle Age"
  else:
    return "Senior"
```

```python
df['age_group'] = df['age'].apply(category)
```

#purchase frequency rate
Purchase frequency = how often a customer buys

It answers:

Who are frequent buyers?
Who are one-time customers?

```python
frequncy_mapping ={
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bimonthly': 60,
    'BiWeekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequncy_mapping)
```
